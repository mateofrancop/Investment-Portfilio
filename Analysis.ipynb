```python
# Calculate daily percentage performance for each closing column
for col in df.columns:
    if col != "Date" and col != "Rendimiento Total":
        df[col + "_Rendimiento (%)"] = (df[col] / df[col].iloc[0] - 1) * 100

# Set the figure size
plt.figure(figsize=(10, 6))

# Iterate through the performance columns and create a line for each one
for col in df.columns:
    if "_Rendimiento (%)" in col:
        sns.lineplot(x="Date", y=col, data=df, label=col)

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Rendimiento (%)")
plt.title("ComparaciÃ³n de Rendimiento de Activos en Porcentaje")
plt.grid(True)

# Show the plot without legend
plt.legend()

# Show the plot
plt.show()

```

```python
# Calculate cumulative total performance
df["Rendimiento Total"] = (df["PETROLEO"] / df["PETROLEO"].iloc[0] - 1) * 100

# Create a total performance chart with Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x="Date", y="Rendimiento Total", data=df, marker="o", color="blue")
plt.xlabel("Date")
plt.ylabel("Rendimiento Total (%)")
plt.title("GrÃ¡fico de Rendimiento Total del Petroleo")
plt.grid(True)

# Show the chart
plt.show()
```

```python
# Calculate cumulative total performance
df["Rendimiento Total"] = (df["S&P 500"] / df["S&P 500"].iloc[0] - 1) * 100

# Create a total performance chart using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x="Date", y="Rendimiento Total", data=df, marker="o", color="blue")
plt.xlabel("Date")
plt.ylabel("Rendimiento Total (%)")
plt.title("GrÃ¡fico de Rendimiento Total del S&P 500")
plt.grid(True)

# Display the chart
plt.show()
```

```python
# Calculate daily percentage returns for all closing columns
for col in df.columns:
    if col != "Date" and col != "S&P 500":
        df[col + "_Rendimiento (%)"] = ((df[col] / df[col].shift(1)) - 1) * 100

# Calculate S&P 500 percentage returns
df["S&P 500_Rendimiento (%)"] = ((df["S&P 500"] / df["S&P 500"].shift(1)) - 1) * 100

# Set the figure size
plt.figure(figsize=(10, 6))

# Iterate through financial asset return columns and create a line for each
for col in df.columns:
    if "_Rendimiento (%)" in col:
        rendimiento_relativo = (df[col] - df["S&P 500_Rendimiento (%)"])
        sns.lineplot(x="Date", y=rendimiento_relativo, label=col, data=df)

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Rendimiento Relativo al S&P 500 (%)")
plt.title("GrÃ¡fico de Rendimiento Relativo al S&P 500 (Diario)")
plt.grid(True)

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python


# Calculate Nasdaq percentage returns
df["NASDAQ_Rendimiento (%)"] = ((df["NASDAQ"] / df["NASDAQ"].iloc[0] - 1)) * 100

# Calculate S&P 500 percentage returns
df["S&P 500_Rendimiento (%)"] = ((df["S&P 500"] / df["S&P 500"].iloc[0] - 1)) * 100

# Set the figure size
plt.figure(figsize=(10, 6))

# Iterate through financial asset return columns and create a line for each
rendimiento_relativo = (df["NASDAQ_Rendimiento (%)"] - df["S&P 500_Rendimiento (%)"])
sns.lineplot(x="Date", y=rendimiento_relativo, label="NASDAQ_Rendimiento (%)", data=df)

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Rendimiento Relativo al S&P 500 (%)")
plt.title("GrÃ¡fico de Rendimiento Relativo al S&P 500 (NASDAQ)")
plt.grid(True)

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
# Calculate Nike percentage returns
df["NIKE_Rendimiento (%)"] = ((df["NIKE"] / df["NIKE"].iloc[0] - 1)) * 100

# Calculate S&P 500 percentage returns
df["S&P 500_Rendimiento (%)"] = ((df["S&P 500"] / df["S&P 500"].iloc[0] - 1)) * 100

# Set the figure size
plt.figure(figsize=(10, 6))

# Iterate through financial asset return columns and create a line for each
rendimiento_relativo = (df["NIKE_Rendimiento (%)"] - df["S&P 500_Rendimiento (%)"])
sns.lineplot(x="Date", y=rendimiento_relativo, label="NIKE_Rendimiento (%)", data=df)

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Rendimiento Relativo al S&P 500 (%)")
plt.title("GrÃ¡fico de Rendimiento Relativo al S&P 500 (NIKE)")
plt.grid(True)

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
# Calculate Microsoft percentage returns
df["MICROFT_Rendimiento (%)"] = ((df["MICROFT"] / df["MICROFT"].iloc[0] - 1)) * 100

# Calculate NASDAQ percentage returns
df["NASDAQ_Rendimiento (%)"] = ((df["NASDAQ"] / df["NASDAQ"].iloc[0] - 1)) * 100

# Set the figure size
plt.figure(figsize=(10, 6))

# Iterate through financial asset return columns and create a line for each
rendimiento_relativo = (df["MICROFT_Rendimiento (%)"] - df["NASDAQ_Rendimiento (%)"])
sns.lineplot(x="Date", y=rendimiento_relativo, label="MICROFT_Rendimiento (%)", data=df)

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Rendimiento Relativo al NASDAQ (%)")
plt.title("GrÃ¡fico de Rendimiento Relativo al NASDAQ (MICROSOFT)")
plt.grid(True)

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
# Calculate daily percentage returns for the financial asset's closing column
df["Rendimiento Diario (%)"] = ((df["APPLE"] / df["APPLE"].shift(1)) - 1) * 100

# Set the Seaborn style (optional)
sns.set(style="whitegrid")

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a histogram of daily returns using Seaborn
sns.histplot(data=df, x="Rendimiento Diario (%)", bins=30, kde=True)

# Customize the plot
plt.xlabel("Rendimiento Diario (%)")
plt.ylabel("Frecuencia")
plt.title("Histograma de Rendimientos Diarios de APPLE")
plt.grid(True)

# Display the chart
plt.show()
```
